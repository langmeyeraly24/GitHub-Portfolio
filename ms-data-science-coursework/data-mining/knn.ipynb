```python
from random import seed
from random import randrange
from csv import reader
from math import sqrt
import pandas as pd
import numpy as np
```

```python
filename = 'phishing_websites.csv'
dataset = pd.read_csv(filename)
dataset = dataset.to_numpy()
print(np.shape(dataset))
```

```python
X = dataset[:, :30]
y = dataset[:, -1]

print(np.shape(X))
print(np.shape(y))
```

```python
print(y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
import time
import matplotlib.pyplot as plt
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"Min: {X.min(axis=0)[:5]}")  
print(f"Max: {X.max(axis=0)[:5]}")

# after scaling
print(f"Mean: {X_scaled.mean(axis=0)[:5]}")
print(f"Std: {X_scaled.std(axis=0)[:5]}")
```

```python
# test k values
k_values = [5, 10, 25, 50, 100, 200]
cv_scores = []
cv_times = []

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for k in k_values:
    start_time = time.time()
    
    knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    scores = cross_val_score(knn, X_scaled, y, cv=cv, scoring='accuracy')
    
    elapsed_time = time.time() - start_time
    
    cv_scores.append(scores.mean())
    cv_times.append(elapsed_time)
    
    print(f"k = {k:3d} | Mean Accuracy: {scores.mean():.4f} (±{scores.std():.4f}) | Time: {elapsed_time:.2f}s")
```
