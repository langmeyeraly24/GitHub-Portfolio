## Type conversions of Variables:
- from string to int
- int to float
- -...

```python
2024 - '2004'
```

```python
# Type conversion:
2024 - int('2004')
```

```python
# Type conversion-1:
birth_year = input ("What is your birth year? ")
age = 2025 - birth_year
print(age)
```

```python
# Type conversion-1:
birth_year = input ("What is your birth year? ")
age = 2025 - int(birth_year)
print(age)
```

```python
# Type conversion-1:
birth_year = int(input ("What is your birth year? "))
age = 2025 - birth_year
print(age)
```
